```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

# 1. Distribution Plots

Plots:
- distplot
- jointplot
- pairplot
- rugplot
- kdeplot
    

```python
df=pd.read_csv('https://assets.datacamp.com/production/course_1392/datasets/gapminder_tidy.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
data = df.dropna(axis='index', how='any')
```

```python
data.info()
```

# displot

```python
# Create displot 
sns.distplot(data['fertility'], bins=10)
```

```python
# remove line on the chart by adding kde=False
sns.distplot(data['fertility'], bins=10, kde=False, color='green')
```


# jointplot

jointplot: scatter, reg, resid, kde, hex

SCATTER PLOT

```python
sns.jointplot(x='child_mortality',y='gdp',data=data, kind ='scatter', color='green')
```

```python
sns.jointplot(x='child_mortality',y='life',data=data, kind ='reg')
```

# pairplot

```python
sns.pairplot(data)
```

# 2. Categorical Plots

- actorplot
- boxplot

- stripplot
- swarmplot

# Bar plot

```python
# load new data 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Create bar plot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# Create countplot
sns.countplot(x='sex',data=tips)
```

```python
HR=pd.read_csv('HR_Data.csv')
```

```python
HR.head()
```

```python
sns.pairplot(HR)
```

# Boxplot

```python
sns.boxplot(x="number_project", y="satisfaction_level", data=HR,palette='rainbow')
```

```python
sns.boxplot(x="number_project", y="satisfaction_level", hue="left",data=HR, palette="coolwarm")
```

# Violin plot

```python
sns.violinplot(x="number_project", y="satisfaction_level", data=HR,hue='left',palette='Set1')
```

# Stripplot

```python
sns.stripplot(x="number_project", y="satisfaction_level", data=HR,hue='left',jitter=True)
```

# Swarmplot

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

# 3. Matrix plot

# Heatmap

```python
sns.heatmap(HR.corr(),cmap='coolwarm',annot=True)
```

Using flight data


```python
flights = sns.load_dataset('flights')
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
```

# Cluster map

clustermap
The clustermap uses hierarchal clustering to produce a clustered version of the heatmap. For example:

```python
sns.clustermap(pvflights)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```
